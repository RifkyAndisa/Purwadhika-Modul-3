```python
#1. Import Library

import pandas as pd
import numpy as np

from sklearn.cluster import KMeans

from scipy.spatial.distance import cdist, pdist 
from sklearn.metrics import silhouette_score
from sklearn.cluster import AgglomerativeClustering

import seaborn as sns
import matplotlib.pyplot as plt
```

# __AGLOMERATIVE__

image.png

```python
#1. Load dataset

df = pd.read_csv('ilustrasi k means.csv', sep = ';')
X = df
X
```

```python
#2. Membuat plot

plt.figure(figsize = (10,10))

sns.set_context(context = 'notebook', font_scale = 2)
sns.scatterplot(
    x = 'tinggibadan',
    y = 'tinggibadanistri',
    data = df)

plt.ylabel('Tinggi badan istri')
plt.xlabel('Tinggi badan')

for i, txt in enumerate(range(0, df.shape[0]-1)):
    plt.annotate(txt, (df['tinggibadan'][i]+0.3, df['tinggibadanistri'][i]+0.3), fontsize = 12)
```

> ### __A. WARD__

```python
#3a. Membuat dendrogram (WARD)

from scipy.cluster.hierarchy import dendrogram, ward, average, complete, single

linkage_array = ward(X) #Membuat linkage

plt.figure(figsize = (10,8))
dendrogram(linkage_array) #Membuat dendrogram

ax = plt.gca()

##buat tambahan
bounds = ax.get_xbound()
ax.plot(bounds, [60,60], '--', c = 'k')
ax.plot(bounds, [30,30], '--', c = 'k')
ax.text(bounds[1], 60, 'two cluster', va = 'center')
ax.text(bounds[1], 30, 'three cluster', va = 'center')
plt.title('Ward')
plt.ylabel('Sample Index')
plt.xlabel('Cluster Distance')

plt.show()
```

> ### __B. AVERAGE__

```python
#3b. Membuat dendrogram (AVERAGE)


linkage_array = average(X) #Membuat linkage

plt.figure(figsize = (10,8))
dendrogram(linkage_array) #Membuat dendrogram

ax = plt.gca()

##buat tambahan
bounds = ax.get_xbound()
ax.plot(bounds, [25,25], '--', c = 'k')
ax.plot(bounds, [15,15], '--', c = 'k')
ax.text(bounds[1], 25, 'two cluster', va = 'center')
ax.text(bounds[1], 15, 'three cluster', va = 'center')
plt.title('Average')
plt.ylabel('Sample Index')
plt.xlabel('Cluster Distance')

plt.show()
```

> ### __C. COMPLETE__

```python
#3c. Membuat dendrogram (COMPLETE)


linkage_array = complete(X) #Membuat linkage

plt.figure(figsize = (10,8))
dendrogram(linkage_array) #Membuat dendrogram

ax = plt.gca()

##buat tambahan
bounds = ax.get_xbound()
ax.plot(bounds, [40,40], '--', c = 'k')
ax.plot(bounds, [30,30], '--', c = 'k')
ax.text(bounds[1], 40, 'two cluster', va = 'center')
ax.text(bounds[1], 30, 'three cluster', va = 'center')
plt.title('Average')
plt.ylabel('Sample Index')
plt.xlabel('Cluster Distance')

plt.show()
```

> ### __D. SINGLE__

```python
#3d. Membuat dendrogram (SINGLE)


linkage_array = single(X) #Membuat linkage

plt.figure(figsize = (10,8))
dendrogram(linkage_array) #Membuat dendrogram

ax = plt.gca()

##buat tambahan
bounds = ax.get_xbound()
ax.plot(bounds, [9, 9], '--', c = 'k')
ax.plot(bounds, [7, 7], '--', c = 'k')
ax.text(bounds[1], 9, 'two cluster', va = 'center')
ax.text(bounds[1], 7, 'three cluster', va = 'center')
plt.title('Single')
plt.ylabel('Sample Index')
plt.xlabel('Cluster Distance')

plt.show()
```

### __RESULT COMPARISON__

```python
agg_ward = AgglomerativeClustering(n_clusters = 3, linkage = 'ward')
df['ward'] = agg_ward.fit_predict(X)

agg_average = AgglomerativeClustering(n_clusters = 3, linkage = 'average')
df['average'] = agg_average.fit_predict(X)

agg_complete = AgglomerativeClustering(n_clusters = 3, linkage = 'complete')
df['complete'] = agg_complete.fit_predict(X)

agg_single = AgglomerativeClustering(n_clusters = 3, linkage = 'single')
df['single'] = agg_single.fit_predict(X)

df
```
